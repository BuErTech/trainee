```python
import urllib
import chardet
from bs4 import BeautifulSoup

response = urllib.request.urlopen('https://movie.douban.com/chart')
html = response.read()

code = chardet.detect(html) #提取编码方式
html = html.decode(code['encoding']) #按编码方式重新编码
print(html)
```

```python
soup = BeautifulSoup(html)  #转为BeautifulSoup格式，HTML格式化，方便提取标签
title = soup.findAll('span', {'style': 'font-size:13px;'}) #提取标签为<span style="font-size:13px;">的字段

for name in title:
    name = str(name)    #转化name为string
    Lcount = name.find('>')    
    Rcount = name.find('</span>')    #判断出左右多余部分的位置
    print(name[Lcount+1:Rcount])   #输出片名
```
