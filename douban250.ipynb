## 豆瓣 250

```python
import requests
from lxml import etree

def page(url):
    urll = "https://movie.douban.com/top250" # url 用到这里，进行抓取的目标地址
    content = requests.get(urll).content                          # 拿下来这个页面的所有的内容
    selector = etree.HTML(content)                                # 把上面的内容转成 Python 可以解析的对象
    result = selector.xpath('//ol[@class="grid_view"]//li')      # 把第三步找到XPath 套取到里面

    for link in result:
        print(link.xpath("./div[@class='item']/div[@class='pic']/em/text()"))
        print(link.xpath("./div[@class='item']/div[@class='pic']/a/img/@src"))
        print(link.xpath("./div[@class='item']/div[@class='info']/div[@class='hd']/a/span[1]/text()"))
        director = (link.xpath("./div[@class='item']/div[@class='info']/div[@class='bd']/p[1]/text()"))

        d1 = (director[0])
        d2 = (director[1])

        d1s = d1.split("主演")

        print(d1s[0].strip())
        #d1s1s = d1s[1].split["//"]
        #print("主演"+d1s1s[0])
        print(d2)
        '''
        \n                            导演: 弗兰克·德拉邦特 Frank Darabont\xa0\xa0\xa0主演: 蒂姆·罗宾斯 Tim Robbins /...', '\n                            1994\xa0/\xa0美国\xa0/\xa0犯罪 剧情\n                        
        '''

    
    
    for i in range(10):
        url = "https://movie.douban.com/top250?start="+str(25*i)
        
        page(url)
```
