```python
import requests

url = "https://dl.lianjia.com/ershoufang/gaoxinyuanqu1/?sug=%E9%AB%98%E6%96%B0%E5%9B%AD%E5%8C%BA"

headers = {
    
    'User-Agent': "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/76.0.3809.100 Safari/537.36",
    
    }

response = requests.request("GET", url, headers=headers)

content=(response.text)
```

```python
selector = etree.HTML(content)                                # 把上面的内容转成 Python 可以解析的对象
# result = selector.xpath('//*[@id="content"]/div/ul/li/div/div/a/@href')      # 把第三步找到XPath 套取到里面
result = selector.xpath('//*[@id="content"]/div/ul/li/div/div/a/text()') 
for link in result:
    print(link)
```
