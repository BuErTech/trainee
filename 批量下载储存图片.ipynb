```python
import requests # 发送请求使用
from urllib.parse import urlencode # 将字典形式的数据转化为查询字符串
import os  # os创建文件，存储爬到的内容
from hashlib import md5  # 生成图片名

def get_page(offset, keyword):
    params = {'offset': offset,
             'format': 'json',
             'keyword': keyword,
             'autoload': 'true',
             'count': '20',
             'cur_tab': '1',
             'from': 'search_tab'} # 从network中查找到data，看query部分
    url = 'https://www.toutiao.com/search_content/?' + urlencode(params)  # 对params进行转码
    header = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/69.0.3497.100 Safari/537.36'}
    try:
        response = requests.get(url, headers = header) # 打开网页
        if response.status_code ==200:  # 200表示成功打开
            return response.json()   # 以json形式返回
    except requests.RequestException:
        print('请求出错')

def get_images(json):
    data = json.get('data')  #取出data中的data部分，放回data
    if data:
        for item in data:  
            image_list = item.get('image_list')   # 取出其中image_list部分，里面存有多个照片信息
            title = item.get('title')   # 取出title部分
            
            if image_list:
                for image in image_list:
                    if type(image) ==dict:
                        yield{
                            'image':image.get('url'),   # 以该种字典形式返回
                            'title': title
                        }
                    else:
                        yield{
                            'image': image,
                            'title':title
                        }

def save_image(item):
    if item.get('title') == []:
        print('搜集完毕')
        return 
    file_path = 'i:\\try\\'+ item.get('title')[0:5] # 照片路径
    if  not os.path.exists(file_path):  # 检查是否有该文件夹，若没有，创建一个
        os.mkdir(file_path)
    
    local_image_url = item.get('image')   # 得到照片url
    response = requests.get('http:' + local_image_url)
    if response.status_code ==200:
        file_name = '{0}{1}.{2}'.format(item.get('title')[0:5],md5(response.content).hexdigest(),'jpg')  # 用format确定文件名
        if not os.path.exists(file_path + '\\'+ file_name):   #查看是否有该文件
            f = open(file_path + '\\'+ file_name, 'wb')  # 以二进制打开文件，若没有，则创建一个
            f.write(response.content) # 到最后结束的时候会报个错

def main(offset, keyword):
    json = get_page(offset, keyword)
    for item in get_images(json):
        print(item)
        save_image(item)

if __name__ == '__main__':
    keyword = '大连'  # 输入关键词
    offset = '5'   # 爬取数目
    main(offset, keyword)
```
