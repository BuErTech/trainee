```python
# -*- coding: utf-8 -*-  
import requests
from lxml import html
from lxml import etree

niukes = list()

# 抓取一个单独的页面
def captureSinglePage(currentPage):
    
    target = "https://www.nowcoder.com/ta/review-frontend/review?query=&asc=true&order=&page="+str(currentPage)
    header = {}
    # 获取页面源码
    r = requests.get(target,headers=header)
    # print(r.text)
    # 解析页面源码
    selector = etree.HTML(r.content.decode('utf-8').encode('utf-8'))
    question = selector.xpath("//div[@class='final-question']")
    print(str(currentPage)+":"+question[0].text)
    #niukes[currentPage]['q'] = ((question[0].text))
    answers = selector.xpath("//div[@class='design-answer-box']//text()")
    # print(lxml.html.tostring(answer[0]).decode('utf-8'))
    answer = ''
    for ans in answers:
        answer+=(ans.encode('utf-8').decode('utf-8'))
        answer+='\n'
    #niukes[currentPage]['a'] = answer
    print(answer)
    
    

for i in range(1,20): 
    captureSinglePage(i)
```
