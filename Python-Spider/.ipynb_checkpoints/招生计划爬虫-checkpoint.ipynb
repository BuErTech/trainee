```python
# -*- coding: utf-8 -*-  
import requests
import csv


from lxml import etree
import xml.etree.cElementTree as ET

target = "http://lnzsks.ooyyee.cn/jh/yx31.html"
header = {"Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3",
"Accept-Encoding": "gzip, deflate",
"Accept-Language": "zh-CN,zh;q=0.9",
"Cache-Control": "no-cache",
"Connection": "keep-alive",
"Host": "lnzsks.ooyyee.cn",
"Pragma": "no-cache",
"Upgrade-Insecure-Requests": "1",
"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.90 Safari/537.36"}

# 获取页面源码
r = requests.get(target,headers=header)


content = str(r.content.decode('utf-8'))



# 解析页面源码
selector = etree.HTML(content)

links = selector.xpath("//a//@href")

# print(len(links))



with open('data.csv', 'a') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['学校招生代码','学校名称','专业招生代码','专业名称','计划','语种','学制','学费'])

for link in links[0:]:
    
    target1 = "http://lnzsks.ooyyee.cn/jh/"+link
    
    # 获取页面源码
    r1 = requests.get(target1,headers=header)
    # 解析页面源码
    # print(r1.status_code)
    if(r1.status_code == 404):
        continue
    
    cotent1 = r1.content.decode(encoding='utf-8')
    
    cotent1 = cotent1.replace("                            "," ")
    cotent1 = cotent1.replace("<font class=yxdd>"," ")
    cotent1 = cotent1.replace("<font class=alt>"," ")
    cotent1 = cotent1.replace("<a name=\"PC_3\"></a>"," ")
    cotent1 = cotent1.replace("<b>","\t")
    cotent1 = cotent1.replace("</b>"," ")
    # cotent1 = cotent1.replace("000","\t000")

    # print(cotent1)

    selector = etree.HTML(cotent1)
    
    
    
    h2 = selector.xpath("//h2")
    
   
    

    columns = selector.xpath("//tr//td/text()")    
    
    index = 0;
    row = list()
    
    column00 = columns[6]
    column01 = columns[7]
    column02 = columns[8]
    row.append(column00)    
    row.append(column01)
    
    for column in columns[12:]:
        index = index+1
        # print((column),end=",")
        
        column = column.encode('utf-8').decode(encoding='utf-8')
        
        if(column.startswith("0")):
            column = column.replace("0","\t0")
       
        row.append(column)
        if(index%6==0):
            with open('data.csv', 'a') as csvfile:
                writer = csv.writer(csvfile)
                
                writer.writerow(row)
                row = list()
                row.append(column00)    
                row.append(column01)
```
