```python
import urllib
from bs4 import BeautifulSoup as bp
import re  # 导入正则表达式

def getHtml(url):   # 获取html对应的soup
    response = urllib.request.urlopen(url)
    soup = bp(response)
    return soup

def findLinks(html):
    # 观察新闻网址组成，用对应的正则表达式概括
    web = html.findAll('a',{'target': '_blank', 'href': re.compile('https:\/\/finance.sina.com.cn.*[0-9].*\.shtml$')})  
    linkList = []
    titleList = []
    dateList = []
    for website in web:
        # 取出链接，保存在linkList中
        linkList.append(website['href'])
        # 取出text，对应的是标题，存在titleList中
        titleList.append(website.get_text())
    for link in linkList:
        # 观察网址组成发现其中有日期，用正则表达式提取
        dateList.append(re.findall(r'[0-9]{4}-[0-9]{2}-[0-9]{2}', link)[0])
    return linkList, titleList, dateList
        
url = 'https://finance.sina.com.cn/'
html = getHtml(url)
links, titles, dates = findLinks(html)

for i in range(0,len(links)):
    print('网址：'+ str(links[i]) + '新闻：'+ str(titles[i]) + '。时间：'+ dates[i])
```
