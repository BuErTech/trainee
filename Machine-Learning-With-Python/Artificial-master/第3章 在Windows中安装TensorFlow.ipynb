# 2.建立TensorFlow的Anaconda虚拟环境  
在一台计算机中，我们常常需要安装很多软件，但是每个软件所需要的Python的关联模块或版本不相同。例如，我们要使用Python开发开发网站系统，安装的网站框架可能需要Python2.X的版本，但是安装Tensorflow需要Python3.5的版本，此时就会发生版本不一致的问题。因为我们需要使用Anaconda虚拟环境来安装，让网站框架与TensorFlow分别在不同的虚拟环境中，这样就不会有版本不一致的问题了。  
## 2.1. 建立工作目录  
在命令提示符串口输入下列命令：  
- 建立并切换到工作目录  
md pythonwork  
cd pythonwork  
执行后屏幕显示界面如图3-8所示：  
![title](images/3-8.png)  

## 2.2.建立Anaconda虚拟环境  
下面使用conda命令建立一个新的Python3.5Anacoda虚拟环境，我们将虚拟环境命名为tensorflow。这个虚拟环境用来安装TensorFlow的版本。在命令提示符窗口输入下列命令：  
- 建立Tensorflow Anaconda虚拟环境  
conda create --name tensorflow python=3.5 anaconda  
执行后屏界面显示如图3-9所示。  
![title](images/3-9.png)   
按Y键之后，就会开始安装Anaconda 虚拟环境，并且安装各个软件包。安装完成后屏幕界面如图3-10所示。  
![title](images/3-10.png)   

## 2.3.启动Anaconda虚拟环境  
建立tensorflow的Anaconda虚拟环境后，就可以启动这个虚拟环境了。在命令提示符窗口输入下列命令：  
- 启动Anaconda虚拟环境  
activate tensorflow  
![title](images/3-11.png)  
命令提示符中出现tensorflow时，表名已经启动tensorflow虚拟环境。  

## 3.在Anaconda虚拟环境安装Tensorflow与Keras  
之前已经建立了Anaconda虚拟环境，现在要在此虚拟环境中安装Tensorflow与Keras  

## 1.启动Anaconda虚拟环境  
安装Tensorflow与keras前，先启动Tensorflow的Anaconda虚拟环境。在命令提示符窗口输入下列命令：  
- 启动Anaconda虚拟环境  
acitvate tensorflow  

## 2.安装Tensorflow  
在命令提示符窗口输入下列命令：  
- 安装Tensorflow命令 pip install tensorflow  
执行后屏幕显示界面如图3-12所示：  
![title](images/3-12.png)  

## 3.安装Keras  
在命令提示符窗口输入下列命令：  
- 安装Tensorflow命令 pip install keras  
执行后屏幕显示界面如图3-13所示：  
![title](images/3-13.png)   


出现上述界面时，表明tensorflow和Keras已经在虚拟环境中安装成功。
