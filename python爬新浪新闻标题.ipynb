```python
import urllib
import chardet
from bs4 import BeautifulSoup

# 对url进行处理，转化为html
def getHtml(url):
    response = urllib.request.urlopen(url)
    html = response.read()
    code = chardet.detect(html)
    html = html.decode(code['encoding'])
    return html

# 取出html中的新闻标题部分
def pickNews(html):
    soup = BeautifulSoup(html)
    newsFirst = soup.findAll('a', {'target':'_blank'}) 
    # 筛选后发现有部分多余的，规律是字数都比较少，拿字数筛选
    news = []
    for name in newsFirst:
        Lcount = str(name).find('>')
        Rcount = str(name).find('<', Lcount)
        if int(Rcount) - int(Lcount) > 10:
            news.append(str(name))
    return news

# 提取出新闻中的网址和标题部分
def splitNews(news):
    websiteAndTitle = []
    for element in news:
        LcountW = element.find('http')
        RcountW = element.find('"',LcountW+1)
        LcountT = element.find('>')
        RcountT = element.find('<', LcountT)
        websiteAndTitle.append('网址：'+ element[LcountW:RcountW]+' 新闻：'+element[LcountT+1:RcountT])
    return websiteAndTitle

html = getHtml('https://news.sina.com.cn/')
news = pickNews(html)
result = splitNews(news)
for new in result:
    print(new)
```
