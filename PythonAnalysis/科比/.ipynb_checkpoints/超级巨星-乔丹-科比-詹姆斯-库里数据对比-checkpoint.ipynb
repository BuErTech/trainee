#  常规数据分析

得分方面,可以说四位巨星都是视得分如探囊取物的人物,在生涯平均得分上,乔神冠绝.

篮板方面,得益于詹姆斯强壮的身体以及出色的运动能力,场均篮板这一项也是领先,细化到进攻篮板和防守篮板,乔神的进攻篮板占总篮板的比例是最高的,这一项湖人名宿科比和乔神是一个档次的,而库里和詹姆斯则是另外一个档次,要落后于科比大概5个百分点,落后乔神大概8个百分点.

助攻方面,这和四位的技术特点是很吻合的,慕斯的场均助攻最多,这也正是我们看骑士比赛可以见到的,慕斯持球推半场,开阔的视野,极高的篮球智商使得慕斯可以把球很舒服地送到每一个合适的球员手中,不过,这貌似也是慕斯被诟病的打法,以及可以部分解释欧文出走波士顿的原因.

抢断和盖帽方面,乔神领先,乔神的防守实在恐怖,3次抢断王,一次最佳防守球员,9次最佳一防,乔神的侵略性怎么吹都不为过.

首发,来看一个有趣的数据,对于出道即巅峰的慕斯而言,整个职业生涯只有一场不是首发,首发率百分之百,这一点上,湖人名宿的值是最低的,这也印证了科比96年进入联盟,兢兢业业从板凳打起,一步一个脚印的成长.

三分球方面,库里一骑绝尘,42%的生涯三分命中率,总命中三分数方面,都已超过其他三位巨星,这一项,库里毫无疑问称王,乔神的三分命中率低,整个职业生涯只命中了572记三分,库里两个赛季都不止这么多,这也体现了联盟的发展,进入小球时代,四分球有望出现啊.

罚球方面,全能如慕斯,这一点和其他三位相比,实在落后颇多,库里再一次称王.

失误方面,慕斯是场均最多的,而乔神再一次耍了流氓.

最后犯规方面,乔科库三者相持平,慕斯是场均犯规数最少的,慕斯在犯规这一项上是控制得最好的.

```python
# 这部分将对四位巨星的得分分布以表格的形式进行展现，使读者对各位巨星的生涯得分分布有个大致印象

import csv
import math
import matplotlib.pyplot as plt
from pandas import Series,DataFrame
with open ('./NBAdata/player_season.csv','r') as csvfile:
    reader = csv.reader(csvfile)
    jordan_score = [row[23] for row in reader if 'Michael Jordan' in row]
with open ('./NBAdata/player_season.csv','r') as csvfile1:
    reader = csv.reader(csvfile1)
    kobe_score = [row[23] for row in reader if 'Kobe Bryant' in row]
with open ('./NBAdata/player_season.csv','r') as csvfile2:
    reader = csv.reader(csvfile2)
    lebron_score = [row[23] for row in reader if 'LeBron James' in row]
with open ('./NBAdata/player_season.csv','r') as csvfile3:
    reader = csv.reader(csvfile3)
    curry_score = [row[23] for row in reader if 'Stephen Curry' in row]
    
jordan_score_process =set(jordan_score)
jordan_set1 = []
jordan_set2 = []
jordan_set3 = []
jordan_set4 = []

kobe_set1 = []
kobe_set2 = []
kobe_set3 = []
kobe_set4 = []

lebron_set1 = []
lebron_set2 = []
lebron_set3 = []
lebron_set4 = []

curry_set1 = []
curry_set2 = []
curry_set3 = []
curry_set4 = []


for i in jordan_score:
    if round(float(i)) < 20:
        jordan_set1.append(i)
    if 20 <= round(float(i)) < 40:
        jordan_set2.append(i)
    if 40 <= round(float(i)) < 60:
        jordan_set3.append(i)
    if round(float(i)) >= 60:
        jordan_set4.append(i)


for i in kobe_score:
    if round(float(i)) < 20:
        kobe_set1.append(i)
    if 20 <= round(float(i)) < 40:
        kobe_set2.append(i)
    if 40 <= round(float(i)) < 60:
        kobe_set3.append(i)
    if round(float(i)) >= 60:
        kobe_set4.append(i)
        
for i in lebron_score:
    if round(float(i)) < 20:
        lebron_set1.append(i)
    if 20 <= round(float(i)) < 40:
        lebron_set2.append(i)
    if 40 <= round(float(i)) < 60:
        lebron_set3.append(i)
    if round(float(i)) >= 60:
        lebron_set4.append(i)
        

for i in curry_score:
    if round(float(i)) < 20:
        curry_set1.append(i)
    if 20 <= round(float(i)) < 40:
        curry_set2.append(i)
    if 40 <= round(float(i)) < 60:
        curry_set3.append(i)
    if round(float(i)) >= 60:
        curry_set4.append(i)
        
data = {'jordan':[len(jordan_set1),len(jordan_set2),len(jordan_set3),len(jordan_set4)],
        'kobe':[len(kobe_set1),len(kobe_set2),len(kobe_set3),len(kobe_set4)],
        'lebron':[len(lebron_set1),len(lebron_set2),len(lebron_set3),len(lebron_set4)],
        'curry':[len(curry_set1),len(curry_set2),len(curry_set3),len(curry_set4)]}
frame = DataFrame(data, columns = ['jordan','kobe','lebron','curry'],
                  index = ['score <= 20','20<score<=40','40<score<=60','score>60'])
frame 
```
