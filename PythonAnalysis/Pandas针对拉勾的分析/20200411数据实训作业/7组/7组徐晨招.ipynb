```python
# 导入pandas库
import pandas as pd
import numpy as np
# 请导入`Job2.csv`文件，并将其命名为`df`
df = pd.read_excel('/home/tianchi/myspace/zhineng.xls')
# 请用`pd.DataFrame.info`的方法查看数据整体信
df.info()
```

```python
# 在`work_loc`这一列中，把`-`替换为空格，统一格式
df['经验要求'] = df['经验要求'].str.replace(' ', '')
#读取df前五行
df.head()
```

```python
# 请选择出我们需要的'salary_at_least','salary_at_most','work_city','exper'4列数据，并保存在df1里
df1 = df[['最低薪资/k','最高薪资/k','工作地点','经验要求']]
# 查看df1是否如我们所愿刚好有且只有我们需要的4列
df1.head()
```

```python
# 查看df1的描述性统计特征
df1.describe()
```

```python
# 请用`pd.Series.agg(['value_counts'])`将数据`df1`中最低薪酬列'salary_at_least'分布的前10行呈现出来
df1['最低薪资/k'].agg(['value_counts']).head(10)
```

```python
# 将`pd.Series.agg(['value_counts'])`将数据`df1`中最高薪酬列'salary_at_most'分布的前10行呈现出来
df1['最高薪资/k'].agg(['value_counts']).head(10)
```

```python
# 查看最低薪资的非重复数据数量
df1['最低薪资/k'].nunique()
```

```python
# 查看最高薪酬的非重复数据数量
df1['最高薪资/k'].nunique()
```

```python
# 需要插入一步数据整理。计算平均薪酬
df1['平均薪资/k'] = (df1['最低薪资/k'] + df1['最高薪资/k'])/2
# 查看df1前5行，验证自己的代码是否正确
df1.head()
```

```python
# 选择'salary_at_mean'这一列，转换数据类型
df1['平均薪资/k'] = df1['平均薪资/k'].astype('int')
df1.head()
```

```python
# 查看平均薪酬的频数分布情况，并将数据赋值为`salary`
salary = df1['平均薪资/k'].agg(['value_counts'])
salary.head()
```
