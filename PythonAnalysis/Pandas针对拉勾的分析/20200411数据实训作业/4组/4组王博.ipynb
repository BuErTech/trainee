```python
math_interval = pd.cut(df_drop['最高薪资/k'],bins=[0,10,20,30,40,100])
```

```python
import matplotlib.pyplot as plt
## 技术类职业最高薪资的分布情况
df_drop.join(math_interval,rsuffix='_interval')
df_drop.join(math_interval,rsuffix='_interval')['最高薪资/k_interval'].value_counts()
df_group_maxsal = df_drop.join(math_interval,rsuffix='_interval')['最高薪资/k_interval'].value_counts()
print(df_group_maxsal)
df_group_maxsal.plot.pie(autopct='%.1f%%',fontsize=20, figsize=(6, 6))

plt.rcParams['font.sans-serif']=['SimHei'] 		#用来正常显示中文标签  
plt.rcParams['axes.unicode_minus']=False

plt.show()
##  在技术类岗位有天赋或是比较努力的同学有10%的机会拿到最高月均超过4w的工资
```

```python
import matplotlib.pyplot as plt
## 技术类职业最低薪资分布状况
math_interval1 = pd.cut(df_drop['最低薪资/k'],bins=[0,10,20,30,40,100])
df_drop.join(math_interval1,rsuffix='_interval')
df_drop.join(math_interval1,rsuffix='_interval')['最低薪资/k_interval'].value_counts()
df_group_minsal = df_drop.join(math_interval1,rsuffix='_interval')['最低薪资/k_interval'].value_counts()
print(df_group_minsal)
df_group_minsal.plot.pie(title = '最低薪资',autopct='%.1f%%',fontsize=20, figsize=(6, 6))

plt.rcParams['font.sans-serif']=['SimHei'] 		#用来正常显示中文标签  
plt.rcParams['axes.unicode_minus']=False

plt.show()
##  在技术类岗位的工作，超过50%都可以拿到月均1w的最低工资
```

以下用的分组+聚合函数

```python
grouped_single = df_drop.groupby('发展阶段')
```

```python
grouped_single = df_drop.groupby('发展阶段')
# 同时使用多个聚合函数
group_max = grouped_single['最高薪资/k']
group_max.agg(['sum','mean','std']).head(13)
```

```python
#  自定义函数
# 官方没有提供极差计算的函数，但通过agg可以容易地实现组内极差计算

grouped_single['最高薪资/k'].agg(lambda x:x.max()-x.min()).head(13)
```

```python
# 判断是否组内最高薪资/k至少有一个值在30-40之间： 
def f(s,low,high):
    return s.between(low,high).mean()

grouped_single['最高薪资/k'].agg(f,30, 40).head(13)
#  技术类工作薪资的天花板非常高
#  除了天使轮的公司外，其他发展阶段的公司有30%的机会拿到月均3-4w的薪资
```

```python
grouped_single[['最高薪资/k','最低薪资/k']].transform(lambda x:(x-x.mean())/x.std()).head()
```

```python
grouped_mul = df_drop.groupby(['工作地点'])
# 指定哪些函数作用哪些列

grouped_mul.agg({'最高薪资/k':['mean','max'],'最低薪资/k':'var'}).sort_values(by='工作地点').tail(15)
```

```python
grouped_mul = df_drop.groupby(['发展阶段','领域'])
# 指定哪些函数作用哪些列

grouped_mul.agg({'最高薪资/k':['mean','max'],'最低薪资/k':'var'})
```
