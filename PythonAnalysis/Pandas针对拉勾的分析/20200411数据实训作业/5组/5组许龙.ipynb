## 进行分组分析

```python
# grouped_1 根据发展阶段分组
grouped_1=df.groupby("发展阶段")
grouped_11=grouped_1["平均薪资/k"]

# grouped_2分析发展阶段与学历要求的关系
grouped_2=df.groupby(["发展阶段","学历要求"])
grouped_22=grouped_2["平均薪资/k"]

# grouped_3根据工作性质分组
grouped_3=df.groupby("工作性质")
grouped_33=grouped_3["平均薪资/k"]

# grouped_4根据工作地点分组
grouped_4=df.groupby("工作地点")
grouped_44=grouped_4["平均薪资/k"]

# grouped_4根据工作的领域分组
grouped_5=df.groupby("领域")
grouped_55=grouped_5["平均薪资/k"]

# grouped_4根据公司的规模分组
grouped_6=df.groupby("公司规模")
grouped_66=grouped_6["平均薪资/k"]
```

```python

from collections import OrderedDict
def f(df):
    data = OrderedDict()
    data['平均薪资_sum'] = df['平均薪资/k'].sum()
    data['最高薪资_var'] = df['最高薪资/k'].var()
    data['最低薪资_mean'] = df['最低薪资/k'].mean()
    return pd.Series(data)
grouped_1.apply(f)
```
