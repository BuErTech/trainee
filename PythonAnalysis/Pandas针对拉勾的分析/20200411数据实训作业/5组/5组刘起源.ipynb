```python
# 用列表可选出多个属性列：

df.groupby(['发展阶段','领域'])["最低薪资/k",'最高薪资/k'].mean()
```

```python
group_mH = grouped_single['最高薪资/k']
group_mH.std().values/np.sqrt(group_mH.count().values)== group_mH.sem().values
```

```python
group_mL = grouped_single['最低薪资/k']
group_mL.std().values/np.sqrt(group_mL.count().values)== group_mL.sem().values
```

```python
group_mL.agg(['sum','mean','std'])
```

```python
grouped_single['最低薪资/k'].agg(lambda x:print(x.head(),'间隔'))
#可以发现，agg函数的传入是分组逐列进行的，有了这个特性就可以做许多事情
```
