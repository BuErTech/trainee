```python
import numpy as np
import pandas as pd
from pyecharts import options as opts
from pyecharts.charts import Pie,Bar,Map

path = '产品.xls'
```

```python
# 不同学历的平均薪资
data = pd.read_excel(path)
data = data[["学历要求","平均薪资/k"]]
data = data.groupby("学历要求").mean().round(1)

data
```

```python
# 产品经理的经验要求
data = pd.read_excel(path)
data = data['经验要求'].value_counts()

data_list = []
count = sum(data.values)
for i,j in data.items():
    data_list.append((i,round(j/count*100,2)))
    
chart = (
    Pie()
    .add('经验要求',data_list)
    .set_global_opts(title_opts=opts.TitleOpts(title="经验要求",pos_top = "5%"))
    .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {c}"+"%"))
).render_notebook()
chart
```

```python
# 不同领域的平均薪资
data = pd.read_excel(path)
data = data[['领域','平均薪资/k']]
data = data.groupby('领域').mean().round(1)
data = sorted(dict(data['平均薪资/k']).items(),key=lambda item:item[1],reverse=True)[:10]

chart= (
    Pie()
    .add('不同领域的平均薪资',data)
    .set_global_opts(title_opts=opts.TitleOpts(title="不同领域的平均薪资",pos_top = "50%"))
    .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}:{c}"+"k"))
).render_notebook()
chart
```

```python
# 产品经理扎堆的前10个领域
from pyecharts.charts import Line
data = pd.read_excel(path)
data = data['领域'].value_counts()
words=sorted(data.items(),key=lambda item:item[1],reverse=True)[:10]
for i in words:
    a=list(i)[0]
    b=list(i)[1]
    print(a)
    print(b)
    
```

```python
from collections import Counter
# 产品经理需求较高的前x个城市
data = pd.read_excel(path)
adress = dict(Counter(data['工作地点']))
adress = sorted(adress.items(),key=lambda item:item[1],reverse=True)[:5]

chart = (
    Pie()
    .add('该城市对产品经理的需求',adress)
    .set_global_opts(title_opts=opts.TitleOpts(title="产品经理需求较高的前5个城市"))
    .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}"))
    
).render_notebook()
chart
```
