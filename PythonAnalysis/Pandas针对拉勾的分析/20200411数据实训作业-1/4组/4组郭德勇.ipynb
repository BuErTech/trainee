```python
df[['工作地点','平均薪资/k','公司规模']].groupby('工作地点').apply(lambda x:x.max())
```

# 基本结论：虽然规模较大的公司平均薪资都普遍较高，但是在一线城市或者发展较好的大城市，规模较小的公司也能提供较高的薪资水平，薪资水平与公司规模有一定的联系，但不是绝对的，表明产品岗的职位有很大的发展前景




```python
import pandas as pd
import numpy  as np
import matplotlib.pyplot as plt
import matplotlib 

df[['工作地点','平均薪资/k']].groupby('工作地点').apply(lambda x:x.max()).plot()
#plt.xticks(rotation=45)    #  设置x轴坐标倾斜度
plt.legend(loc = "best")
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.show()
```

# 基本结论：一线城市的平均薪资都较高，在大城市中选择就业会有更好的发展

```python
import pandas as pd
import numpy  as np
import matplotlib.pyplot as plt
import matplotlib 
grouped_single = df.groupby('经验要求')
df[['经验要求','平均薪资/k']].groupby('经验要求').apply(lambda x:x.max()).plot()
plt.xticks(rotation=45)    #  设置x轴坐标倾斜度
plt.legend(loc = "best")
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.show()
```

# 基本结论：产品岗工作经验5-10年的平均薪资达到100k以上，而对于应届毕业生，平均薪资也在20k左右，薪资起点较高
