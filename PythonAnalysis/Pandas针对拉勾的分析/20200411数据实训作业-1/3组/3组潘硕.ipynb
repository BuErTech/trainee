# 单级索引
#### 1. loc方法、iloc方法、[]操作符
最常用的索引方法可能就是这三类，其中iloc表示位置索引，loc表示标签索引，[]也具有很大的便利性，各有特点

##### （a）loc方法

```python
#单行索引（大圣科技）
df.loc[4878943]
```

```python
# 多行索引（大圣科技，虎牙信息）
df.loc[[4878943,6210428]]
```

```python
df.loc[:4878943:].head()
```

```python
#单列索引

df.loc[:,'发展阶段'].head()
```

```python
#单列索引

df.loc[:,['发展阶段',"需求匹配"]].head()
```

```python
#联合索引
df.loc[4878943:6210428,'发展阶段':'需求匹配'].head()
```

```python
#函数式索引

df.loc[lambda x:x['发展阶段']=='上市公司'].head()
```

```python
df.loc[df["公司规模"].isin(["2000人以上"])].head()
```
