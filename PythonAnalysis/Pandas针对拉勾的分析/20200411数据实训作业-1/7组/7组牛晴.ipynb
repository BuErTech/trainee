```python
#在移动互联网-电商领域，工作地为北京的企业中，企业的产品经理的最低薪资从高到低排列
import numpy as np
import pandas as pd
df = pd.read_excel('/home/tianchi/myspace/产品.xls',index_col='平均薪资/k')
df1=df[['领域','工作地点','公司名称','最高薪资/k','最低薪资/k']] #这几列是我想要的
df1.sort_values(by='最低薪资/k',ascending=False) # 将最高薪资按倒序排列
grouped_mul = df1.groupby(['领域','工作地点'])#按照领域和工作地点分组
grouped_mul.get_group(('移动互联网,电商','北京'))#只看北京的电商领域的公司薪资
```

```python
#每个领域的的不同工作地点之间的最低薪资对比
import numpy as np
import pandas as pd
df = pd.read_excel('/home/tianchi/myspace/产品.xls')
df1=df[['领域','工作地点','公司名称','最高薪资/k','平均薪资/k','最低薪资/k']] 
#df1.sort_values(by='最高薪资/k',ascending=False) 
group1=df1.groupby(['领域','工作地点'])
group1.agg({'最低薪资/k':['max']})
```
