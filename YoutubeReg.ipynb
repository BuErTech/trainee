```python
file = open("./youtube.txt", "r") 

content = file.read()
```

```python
import re 


pattern = re.compile(r'''<span id="video-title" class="style-scope ytd-playlist-video-renderer" aria-label="(.*)title="(.*)">''')

results = re.match(pattern, content)
```

```python
print(results)
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(content,"html5lib")


print(soup.prettify())
```

```python
titles = soup.find_all(attrs={"id": "video-title"})

print(len(titles))
```

```python
titles[0].string
```

```python
index = 1
for t in titles:
    temp = t.string
    temp = temp.replace("\n","")
    print(str(index)+","+temp.strip())
    index = index + 1
```
